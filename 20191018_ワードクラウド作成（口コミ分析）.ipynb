```python
import sys
import MeCab

def mecab_list(text):
    #https://qiita.com/menon/items/2b5ad487a98882289567
    tagger = MeCab.Tagger()
    tagger.parse('')
    node = tagger.parseToNode(text)
    word_class = []
    while node:
        word = node.surface
        wclass = node.feature.split(',')
        if wclass[1] in ["形容動詞語幹","サ変接続","一般","固有名詞","ナイ形容詞語幹"]:
        #if wclass[0] != u'BOS/EOS':
            if wclass[0] in ["副詞","記号"]:
                pass
            elif wclass[6] == None:
                #word_class.append((word,wclass[0],wclass[1],wclass[5],""))
                word_class.append(word)
            else:
                #word_class.append((word,wclass[0],wclass[1],wclass[4],wclass[5],wclass[6]))
                word_class.append(wclass[6])
        node = node.next
    return word_class

f=open("C:/Users/tong/Desktop/201906_IPHONE/作業台/bad.txt","r")
fp=f.read()

wordlist=mecab_list(fp)
wordlist
```

```python
#Python - Word Cloud を作成する方法について - Pynote
#http://pynote.hatenablog.com/entry/python-wordcloud
from collections import Counter 
import matplotlib.pyplot as plt
import wordcloud
import numpy as np
from PIL import Image
mask = np.array(Image.open("C:/Users/tong/Desktop/201906_IPHONE/作業台/d3.png"))
cloudlist=Counter(wordlist)
cloud = wordcloud.WordCloud(font_path="C:\Windows\Fonts\meiryo.ttc",width=300, height=500,background_color="white",contour_width=3,contour_color='steelblue',mask=mask).fit_words(cloudlist)
plt.figure(figsize=(10,10))
plt.imshow(cloud) # 在坐标轴上显示图像
plt.axis("off") # 去除图像坐标轴
plt.show() # 显示词云图
```
