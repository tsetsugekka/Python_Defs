```python
#重回帰分析
#https://pythondatascience.plavox.info/scikit-learn/%E7%B7%9A%E5%BD%A2%E5%9B%9E%E5%B8%B0

from sklearn import linear_model
clf = linear_model.LinearRegression()

# 説明変数
wine_except_quality = data.drop("iPhone修理の全般的な満足度", axis=1)
X = wine_except_quality.as_matrix()

# 目的変数
Y = data['iPhone修理の全般的な満足度'].as_matrix()

# 予測モデルを作成
clf.fit(X, Y)
 
# 偏回帰係数
print(pd.DataFrame({"Name":wine_except_quality.columns,
                    "Coefficients":clf.coef_}).sort_values(by='Coefficients') )
 
# 切片 (誤差)
print(clf.intercept_)
```

```python

 
# データフレームの各列を正規化
wine2 = data.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
wine2.head()
 
# 説明変数に "quality (品質スコア以外すべて)" を利用
wine2_except_quality = wine2.drop(["iPhone修理の全般的な満足度","問診・対応が始まるまでの所要時間"], axis=1)
X = wine2_except_quality.as_matrix()
 
# 目的変数に "quality (品質スコア)" を利用
Y = wine2['iPhone修理の全般的な満足度'].as_matrix()
 
# 予測モデルを作成
clf.fit(X, Y)
 
# 偏回帰係数
print(pd.DataFrame({"Name":wine2_except_quality.columns,
                    "Coefficients":np.abs(clf.coef_)}).sort_values(by='Coefficients') )
 
# 切片 (誤差)
print(clf.intercept_)

'''
[iPhone修理の全般的な満足度]=
0.06×（修理サービスが完了するまでの所要時間）
0.55×（問題の解決）＋
0.18×（ショップ担当者からの説明・接客対応）＋
0.05×（ショップ担当者の問題を診断する技能）＋
0.12×（即日修理オプションの説明）
'''
```
